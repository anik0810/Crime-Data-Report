```python
import pandas as pd
df=pd.read_csv('crime.csv')
```

```python
df
```

```python
data_andhra=df.loc[df['STATE/UT']=='ANDHRA PRADESH']
data_andhra=data_andhra.loc[df['YEAR']==2001]
```

```python
data_andhra
```

```python
data_andhra_2001=data_andhra.drop(['STATE/UT','DISTRICT','YEAR','TOTAL IPC CRIMES'],1)
# data_andhra.drop(['YEAR'],1)
```

```python
data_sum=data_andhra_2001.sum()
data_sum
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (15, 5))
data_sum.plot(kind='bar',width=0.6,color='maroon')
```

```python
print(f"The most occured crime is HURT/GREVIOUS HURT and the number is {data_sum.max()} in 2001")
```
